```python
# Question 11: Find the most frequent character in "mississippi"

char = "mississippi"
freq = {}
for ch in char:
    freq[ch] = freq.get(ch,0)+1
    max_count = max(freq.values())
for ch in freq:
    if freq[ch] == max_count:
        print(ch)
```

```python
# Question 23: Convert "hello world" to "hElLo WoRlD" (alternating case)

input = "hello world"
result = ""

for i in range(len(input)):
    if i % 2 == 0:
        result += input[i].lower()
    else:
        result += input[i].upper()

print(result)
```

```python
# Question 30: Reverse words in "Hello World Python"

text = "Hello World Python"

words = text.split()
result = ""

for i in range(len(words)-1, -1, -1):
    result += words[i] + " "

print(result.strip())
```
