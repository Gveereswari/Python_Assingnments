```python
# Question 1: Check if 25 is greater than 20 and less than 30

number = 25
is_greater_than_20 = number > 20
is_less_than_30 = number < 30
print(f"{number} is greater than 20 :{is_greater_than_20}")
print(f"{number} is lesser than 30 :{is_less_than_30} ")
```
