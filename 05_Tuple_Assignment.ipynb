```python
# Question 5: Count how many times 5 appears in (1, 5, 2, 5, 3, 5, 4, 5, 6)
tup=(1,5, 2, 5, 3, 5, 4, 5, 6)
count=0
for i in tup:
    if i==5:
        count+=1
print(count) 
```

```python
# Question 8: Check if 25 exists in tuple (10, 20, 30, 40, 50)

N=(10, 20, 30, 40, 50)
for i in N:
    if i==25:
        print("Exists")
        break
else:
    print("Does not exist")
```

```python
# Question 9: Create a tuple of first 5 even numbers

#way 1:
for i in range (2,11,2):
    print(i) 

#way 2:
even=([i for i in range(2,11) if i%2==0])
print(even)

```

```python
# Question 10: Find the average of numbers in tuple (15, 23, 31, 42, 56, 78)

n=(15, 23, 31, 42, 56, 78)
for i in n:
    total=sum(n)
    avg=total/len(n)
print(avg)  
```
